```python
import pandas as pd
import numpy as np

# read tax_docscores file
df_score = pd.read_pickle("../data/tax_docscores_nl.pkl")
df_score.drop_duplicates(inplace=True)
df_score
```

```python
# Create a column based on score value (1for >= 0.5 and 0 for <= 0.5) and assign dummies(0,1) for the tax
df_score["Tax"] = np.where(df_score['docscores'] > 0.5, 1, 0)
df_score.head(30)
```

```python
# count tax or non tax documents avaleble in datadrame 
df_score["Tax"].value_counts() 
```

```python
# load Final-File
df1= pd.read_pickle("../data/Final-File.pkl")
#df1.drop_duplicates(inplace=True)
#df1["Final-Text"]
#df1.shape
df1.head(10)
```

```python
# Join above df_score and df dataframe
df_all = pd.concat([df1, df_score], axis=1, join='inner')
df_all.head(10)
#df_all.shape
```

```python
# create df for tax file
df_tax=df_all[df_all["Tax"]==1]
df_tax.shape
```

```python
# create df for non-tax file
df_Non_tax=df_all[df_all["Tax"]==0]
df_Non_tax.shape
```

```python
# use downsaml method
df_Non_tax_downsampled =df_Non_tax.sample(df_tax.shape[0])
df_Non_tax_downsampled.shape
```

```python
# create balance data set
df_balanced=pd.concat([df_tax,df_Non_tax_downsampled])
df_balanced.shape
```

```python
df_balanced['Tax'].value_counts()
```

```python
df_balanced.sample(5)
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier

df_balanced = df_balanced[df_balanced['Vector'].map(len) > 0]
X = np.array([np.array(elm) for elm in df_balanced.Vector.tolist()])
y=df_balanced['Tax']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
```

```python
X_train
```

```python
 # define the model
model = RandomForestClassifier(n_estimators = 300, max_depth = 7, random_state =42, class_weight= "balanced_subsample")
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
from sklearn import metrics

metrics.plot_confusion_matrix(model, X_test, y_test,display_labels=['Tax-File', 'Non-Tax-File'],)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from numpy import mean

# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(model, X, y, scoring='f1_micro', cv=cv, n_jobs=-1)
score = mean(scores)
print('F1 Score: %.3f' % score)
```
