```python
import pandas as pd
import numpy as np
df = pd.read_pickle("../data/tax_keywords_nl.pkl")
df
```

```python
# read tax_docscores file
df_score = pd.read_pickle("../data/tax_docscores_nl.pkl")
df_score.drop_duplicates(inplace=True)
df_score
```

```python
# round off the columns
df_score.round({"docscores":1}).head(10)
```

```python
# Create a column based on score value (1for >= 0.5 and 0 for <= 0.5) and assign dummies(0,1) for the tax
df_score["Tax"] = np.where(df_score['docscores'] > 0.5, 1, 0)
df_score.head(30)
```

```python
# count tax or non tax documents avaleble in datadrame 
df_score["Tax"].value_counts() 
```

```python
# load Final-File
df1= pd.read_pickle("../data/Final-File.pkl")
#df1.drop_duplicates(inplace=True)
#df1["Final-Text"]
#df1.shape
df1.head(10)
```

```python
# Join above df_score and df dataframe
df_all = pd.concat([df1, df_score], axis=1, join='inner')
df_all.head(10)
#df_all.shape
```

```python
#266/660
```

```python
# create df for tax file
df_tax=df_all[df_all["Tax"]==1]
df_tax.shape
```

```python
# create df for non-tax file
df_Non_tax=df_all[df_all["Tax"]==0]
df_Non_tax.shape
```

```python
# use downsaml method
df_Non_tax_downsampled =df_Non_tax.sample(df_tax.shape[0])
df_Non_tax_downsampled.shape
```

```python
# create balance data set
df_balanced=pd.concat([df_tax,df_Non_tax_downsampled])
df_balanced.shape
```

```python
df_balanced['Tax'].value_counts()
```

```python
df_balanced.sample(5)
```

```python
# To train & test data set invoce sklearn and split dataset
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(df_balanced["Final_text"],df_balanced["Tax"], stratify=df_balanced["Tax"])
```

```python
X_train.head(5)
```

```python
# import tensorflow and Bert model

import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text

bert_preprocess = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
bert_encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4")
```

```python
# Some example of embeding
def get_sentence_embeding(sentences):
    preprocessed_text = bert_preprocess(sentences)
    return bert_encoder(preprocessed_text)['pooled_output']

get_sentence_embeding(["I love to eat fish", "my daughter name is Aria"])
```

```python
e=get_sentence_embeding(["apple","banana","mango" "Sunday","Monday","Tuesday"])
```

```python
# use cosine-similarity method
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity([e[0]],[e[4]])
```

```python
# Bert Layers

text_input=tf.keras.layers.Input(shape=(),dtype=tf.string,name="text")
preprocessed_text=bert_preprocess(text_input)
outputs=bert_encoder(preprocessed_text)

# Neural network Layers

e = tf.keras.layers.Dropout(0.1,name="dropout")(outputs['pooled_output'])
e=tf.keras.layers.Dense(1, activation="sigmoid", name='output')(e)

# construct final model
model=tf.keras.Model(inputs=[text_input], outputs=[e])
```

```python
model.summary()
```

```python
METRICS = [tf.keras.metrics.BinaryCrossentropy(name='accuracy'),
           tf.keras.metrics.Precision(name='precision'),
           tf.keras.metrics.Recall(name='recall')]

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=METRICS)
```

```python
model.fit(X_train,y_train,epochs=5)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_predicted=model.predict(X_test)
y_predicted=y_predicted.flatten()
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test,y_predicted)
cm
```
